```python
from ultralytics import YOLO

model = YOLO("/kaggle/working/soccer-video-analytics/models/player_tracking_model.pt")

results = model.train(
    # --- Core Parameters ---
    data="/kaggle/working/soccer-1/data.yaml",
    epochs=500,
    patience=50,              # Stop training if no improvement is seen for 50 epochs
    imgsz=1280,               # Keep high-res: CRITICAL for small object detection
    batch=4,                 # Auto-batch size is great for maximizing GPU utilization
    device="0,1",
    lr0=0.0001,
    mosaic=1.0,               # Excellent for varying ball scales and context
    mixup=0.0,                # For single-class detection (just 'ball'), mixup can be less effective. Better to disable.
    copy_paste=0.1,           # Increased slightly. Very powerful for small objects, simulates more balls on screen.
    degrees=180,               # Increased slightly for more camera rotation variety
    translate=0.1,
    scale=0.5,                # Aggressive scaling is good for simulating camera zooms
    shear=5,                  # Slight increase in shear
    perspective=0.0001,        # Add slight perspective shifts, common in broadcasts

    # Color Augmentations
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    amp=True,           # ✅ AMP DISABLED
    # Other Important Parameters
    fliplr=0.5,               # Horizontal flips are fine for football
    multi_scale=True,         # Train on varying image sizes for scale invariance
    cos_lr=True,              # Use cosine learning rate scheduler for smoother convergence
    optimizer='auto',        # AdamW is often a robust choice
    
    # --- Experiment Management ---
    project='football-tracker', # Organize all runs in this folder
    name='yolov8s_run_1280px',  # Give this specific run a descriptive name
)
```
